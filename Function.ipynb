01.  What is the difference between a function and a method in Python?

Ans:- In Python, the difference between a function and a
      method is based on the context in which they are used:

     Function: A function is a block of reusable code that can be called directly. It is defined using the def keyword and can be called independently.

     Method: A method is a function that is associated with an object. It is a function that belongs to a class, and it is called on an instance of that class (or sometimes the class itself, in the case of class methods).  

     Example:

     i.Function:
                
              def greet(name):
                 return f"Hello, {name}!"

              # Calling the function
              print(greet("Alice"))  # Output: Hello, Alice!

     02.Method:

               class Person:
                  def __init__(self, name):
                     self.name = name
    
               def greet(self):
                 return f"Hello, {self.name}!"

        # Creating an instance of Person
          person = Person("Alice")

        # Calling the method on the instance
          print(person.greet())  # Output: Hello, Alice!

05. What are iterators in Python and how do they differ

Ans:- In Python:

Key Difference:
Iterable: Can be iterated over but does not hold the state of iteration.

Iterator: Keeps track of the iteration state and provides the next item.

Example:

06. Explain the concept of generators in Python and how
    they are defined

Ans:- Generators in Python are a type of iterable that allow you to iterate over a sequence of values lazily, meaning values are produced one at a time as needed, rather than all at once. They are defined using a function with the yield keyword.

Key Points:
Generators use yield instead of return.

They produce values on the fly, saving memory.

Example:

         def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

# Using the generator

gen = count_up_to(5)

for number in gen:

    print(number)

Output:

 1
 2
 3
 4
 5


07.  What are the advantages of using generators over
     regular functions?

Ans:- Advantages of using generators over regular functions:

Memory Efficiency: Generators generate values one at a time, so they donâ€™t store all values in memory at once.

Lazy Evaluation: Values are produced only when needed, improving performance for large data sets.

State Retention: Generators maintain their state between yield calls, allowing them to resume from where they left off.

Example:

    Without generator (using a list):

     def get_squares(n):
    return [x*x for x in range(1, n+1)]

squares = get_squares(1000000)  # This consumes a lot of memory

    With generator:

          def get_squares(n):
           for x in range(1, n+1):
                 yield x*x

squares = get_squares(1000000)  # More memory efficient

10. What is the difference between `map()`, `reduce()`,
    and `filter()` functions in Python?

Ans:- In Python, map(), reduce(), and filter() are functional programming tools used for processing iterables, but they serve different purposes:

01.map():

Applies a given function to each item in an iterable and returns an iterable of the results.

Purpose: Transformation of data.

02.reduce() (from functools):

Applies a binary function (a function that takes two arguments) cumulatively to the items of an iterable, reducing it to a single value.

Purpose: Reduces data to a single result.

03.filter():

Applies a function to each item in an iterable and returns an iterable of items for which the function returns True.

Purpose: Filtering elements based on a condition.

Example:    

        from functools import reduce

# map: squares each number
numbers = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, numbers)  # Output: [1, 4, 9, 16]

# reduce: sums all numbers
sum_of_numbers = reduce(lambda x, y: x + y, numbers)  # Output: 10

# filter: filters even numbers
even_numbers = filter(lambda x: x % 2 == 0, numbers)  # Output: [2, 4]

print(list(squared))  # Output: [1, 4, 9, 16]
print(sum_of_numbers)  # Output: 10
print(list(even_numbers))  # Output: [2, 4]

04. Write a Python function that checks if a given number
    is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):  # Check divisibility up to the square root of n
        if n % i == 0:
            return False
    return True

# Example usage
for num in range(1, 201):
    if is_prime(num):
        print(num)  # Prints prime numbers from 1 to 200
```

05.  Create an iterator class in Python that generates  
     the Fibonacci sequence up to a specified number of
     terms.

```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms  # The number of terms to generate
        self.a, self.b = 0, 1  # Starting values for the Fibonacci sequence
        self.count = 0  # Counter to track the number of terms generated

    def __iter__(self):
        return self  # Return the iterator object itself

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration  # Stop the iteration if the specified number of terms is reached
        fib_number = self.a
        self.a, self.b = self.b, self.a + self.b  # Update the Fibonacci sequence
        self.count += 1
        return fib_number

# Example usage
fibonacci = FibonacciIterator(10)  # Generate the first 10 Fibonacci numbers
for num in fibonacci:
    print(num)
```

06. Write a generator function in Python that yields the
    powers of 2 up to a given exponent.

```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage
for power in powers_of_two(5):
    print(power)
```

09.  Write a Python program that uses `map()` to convert  
     a list of temperatures from Celsius to Fahrenheit.

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 25, 30, 100]

# Using map to apply the conversion function to each element in the list
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Output the converted temperatures
print(fahrenheit_temperatures)
```

10. Create a Python program that uses `filter()` to
     remove all the vowels from a given string.

```python
# Function to check if a character is not a vowel
def is_not_vowel(char):
    vowels = "aeiouAEIOU"
    return char not in vowels

# Given string
input_string = "Hello, World!"

# Using filter to remove vowels
filtered_string = ''.join(filter(is_not_vowel, input_string))

# Output the filtered string
print(filtered_string)
```
